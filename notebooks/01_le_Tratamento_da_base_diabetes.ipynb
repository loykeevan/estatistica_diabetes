```python
import pandas as pd

# funções desenvolvidas para o projeto
from src.config import DADOS_ORIGINAIS
```

```python
df_diabetes = pd.read_csv(DADOS_ORIGINAIS)
df_diabetes.head()
```

```python
df_diabetes.info()
```

Todas as colunas da base de dados acima são classificados como float quando na verdade, algumas dessas colunas podem ser classificadas como categóricas.
Para uma melhor visualização dos dados, vizualizaremos as 5 primeiras linhas da base de dados por completo.

```python
with pd.option_context("display.max_columns", None):
    display(df_diabetes.head())
```

Para uma melhor compreensão dos dados foram mudados os nomes das colunas

```python
df_diabetes.columns = [
    "Diabetes",
    "PressaoAlta",
    "ColesterolAlto",
    "ColesterolExame",
    "IMC",
    "Fumante",
    "AVC",
    "ProblemaCardiaco",
    "AtividadeFísica",
    "ConsumoFrutas",
    "ConsumoVegetais",
    "ConsumoBebidaAlcoolica",
    "PlanoSaude",
    "SemDinheiroConsulta", 
    "SaudeGeral",
    "SaudeMental",
    "SaudeFisica",
    "DificuldadeLocomocao", 
    "Genero", 
    "Idade",
    "Educacao",
    "FaixaRenda",
    ]
```

```python
with pd.option_context("display.max_columns", None):
    display(df_diabetes.head())
```

Orgranizando as colunas de acordo com os tipos de dados que elas possuem:

```python
colunas_binarias = df_diabetes.nunique()[df_diabetes.nunique() == 2].index.to_list()
colunas_binarias
```

A partir desse ponto, com o intuito de manter a base de dados íntegra, será criada uma cópia da base de dados para melhores tratamentos

```python
df_diabetes_tratado = df_diabetes.copy()
```

Com as colunas binárias determinadas, vamos mudar os tipos de dados e de cada uma dessas colunas. Além disso, para uma melhor compreensão dos dados, vamos mudar os tipos binários de cada coluna para "Não" e "Sim", exceto a coluna "Genero"  

```python
for coluna in colunas_binarias:
    if coluna != "Genero":
        df_diabetes_tratado[coluna] = pd.Categorical(df_diabetes_tratado[coluna]).rename_categories(["Não", "Sim"])
    else:
        df_diabetes_tratado[coluna] = pd.Categorical(df_diabetes_tratado[coluna]).rename_categories(["Feminino", "Masculino"])
```

```python
with pd.option_context("display.max_columns", None):
    display(df_diabetes_tratado.head())
```

```python
df_diabetes_tratado.info()
```

Do mesmo modo que as colunas binárias foram modificadas para colunas categóricas, modificaremos as demais colunas para o mesmo tipo categórico

```python
df_diabetes_tratado["SaudeGeral"] = pd.Categorical(
    df_diabetes_tratado["SaudeGeral"], 
    ordered=True,
    ).rename_categories(["Excelent", "Muito Boa", "Boa", "Aceitável", "Ruim"])
```

```python
df_diabetes_tratado["Idade"] = pd.Categorical(
    df_diabetes_tratado["Idade"], 
    ordered=True,
    ).rename_categories([
        "18-24", 
        "25-29",
        "30-34",
        "35-39", 
        "40-44", 
        "45-49", 
        "50-54", 
        "55-59", 
        "60-64",
        "65-69", 
        "70-74", 
        "75-79", 
        "80+"
    ])
```

```python
df_diabetes_tratado["Educacao"] = pd.Categorical(
    df_diabetes_tratado["Educacao"], 
    ordered=True,
    ).rename_categories([
        "sem estudo", 
        "primario",
        "secundario inc.",
        "secundario", 
        "faculdade inc. tecnico", 
        "faculdade", 
    ])
```

```python
df_diabetes_tratado["FaixaRenda"] = pd.Categorical(
    df_diabetes_tratado["FaixaRenda"], 
    ordered=True,
    ).rename_categories([
        "< 10.000",
        "10.000-14.999",
        "15.000-19.999",
        "20.000-24.999",
        "25.000-34.999",
        "35.000-49.999",
        "50.000-74.999",
        "75.000+",
    ])
```

```python
df_diabetes_tratado.info()
```

Já as colunas numéricas, há a necessidade de fazer uma análise para saber se todas os dados se tratam de números inteiros.

```python
df_diabetes_tratado["IMC"].apply(float.is_integer).all()
```

```python
df_diabetes_tratado["SaudeMental"].apply(float.is_integer).all()
```

```python
df_diabetes_tratado["SaudeFisica"].apply(float.is_integer).all()
```

Como todos os dados se tratam de números inteiros, mudaremos o tipo de dado (de floar para inteiro) para obtermos menor espaço de memória e maior performance

```python
colunas_numericas = df_diabetes_tratado.select_dtypes(include="number").columns.to_list()
colunas_numericas
```

```python
for coluna in colunas_numericas:
    df_diabetes_tratado[coluna] = pd.to_numeric(
        df_diabetes_tratado[coluna],
        downcast="integer"
    )
```

```python
df_diabetes_tratado.info()
```
